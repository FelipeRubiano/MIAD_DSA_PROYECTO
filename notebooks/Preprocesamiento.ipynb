```python
import pandas as pd
import plotly.express as px
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
df = pd.read_csv("../data/processed/Bank_Customer_Churn.csv")
```

```python
# Preprocesamiento -----------------------------------------------------------------------------------

df = df.copy()

# Codificar variables categóricas
cat_cols = ['country', 'gender']
df[cat_cols] = df[cat_cols].apply(LabelEncoder().fit_transform)

# Separar variables predictoras y target
X = df.drop(columns=['customer_id', 'churn'])
y = df['churn']

# Escalar las variables numéricas (opcional, mejora estabilidad)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir en train y test
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# Convertir los arrays a DataFrames
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)
y_train = pd.Series(y_train, name="Exited")
y_test = pd.Series(y_test, name="Exited")
```

```python
X_train.to_csv("../data/processed/X_train.csv", index=False)
X_test.to_csv("../data/processed/X_test.csv", index=False)
y_test.to_csv("../data/processed/y_test.csv", index=False)
y_train.to_csv("../data/processed/y_train.csv", index=False)
```
